```python
# ---------------- STORAGE ENGINE ----------------

class StorageEngine:

    def create_table(self, table_name):
        file_name = table_name + ".txt"
        try:
            f = open(file_name, "x")
            f.close()
            print("Table created successfully")
        except FileExistsError:
            print("Table already exists")

    def insert(self, table, values):
        file_name = table + ".txt"
        try:
            f = open(file_name, "a")
            row = ",".join(values)
            f.write(row + "\n")
            f.close()
            print("Record inserted")
        except FileNotFoundError:
            print("Table does not exist")

    def select(self, table):
        file_name = table + ".txt"
        try:
            f = open(file_name, "r")
            data = f.readlines()
            f.close()

            print("\n--- Records ---")
            if len(data) == 0:
                print("(empty)")
            else:
                for row in data:
                    print(row.strip())

        except FileNotFoundError:
            print("Table not found")

    def delete(self, table):
        file_name = table + ".txt"
        try:
            f = open(file_name, "w")  # clear file
            f.close()
            print("All records deleted")
        except FileNotFoundError:
            print("Table not found")


# ---------------- PARSER ----------------

class Parser:

    def parse(self, query):
        query = query.strip()

        if not query:
            return ("ERROR", "Empty Query")

        # CREATE TABLE
        if query.upper().startswith("CREATE"):
            tokens = query.split()
            if len(tokens) >= 3:
                return ("CREATE", tokens[2])
            else:
                return ("ERROR", "Invalid CREATE syntax")

        # INSERT INTO
        elif query.upper().startswith("INSERT"):
            try:
                parts = query.split("VALUES")
                table = parts[0].split()[2]

                values = parts[1].strip()
                values = values.replace("(", "").replace(")", "")
                values = values.split(",")
                values = [v.strip() for v in values]

                return ("INSERT", table, values)
            except:
                return ("ERROR", "Invalid INSERT syntax")

        # SELECT
        elif query.upper().startswith("SELECT"):
            tokens = query.split()
            if len(tokens) >= 4:
                return ("SELECT", tokens[-1])
            else:
                return ("ERROR", "Invalid SELECT syntax")

        # DELETE
        elif query.upper().startswith("DELETE"):
            tokens = query.split()
            if len(tokens) >= 3:
                return ("DELETE", tokens[2])
            else:
                return ("ERROR", "Invalid DELETE syntax")

        else:
            return ("ERROR", "Invalid Query")


# ---------------- EXECUTOR ----------------

class Executor:

    def __init__(self):
        self.storage = StorageEngine()

    def execute(self, parsed_query):

        command = parsed_query[0]

        if command == "CREATE":
            self.storage.create_table(parsed_query[1])

        elif command == "INSERT":
            self.storage.insert(parsed_query[1], parsed_query[2])

        elif command == "SELECT":
            self.storage.select(parsed_query[1])

        elif command == "DELETE":
            self.storage.delete(parsed_query[1])

        elif command == "ERROR":
            print("Error:", parsed_query[1])

        else:
            print("Invalid Query")


# ---------------- CLI ----------------

class CLI:

    def start(self):

        parser = Parser()
        executor = Executor()

        print("MiniDB Engine Started")
        print("Commands: CREATE, INSERT, SELECT, DELETE")
        print("Type EXIT to quit\n")

        while True:
            query = input("BuildDB> ")

            if query.upper() == "EXIT":
                print("Exiting BuildDB...")
                break

            parsed = parser.parse(query)
            executor.execute(parsed)


# ---------------- MAIN ----------------

if __name__ == "__main__":
    cli = CLI()
    cli.start()
```
